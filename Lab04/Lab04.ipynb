```python
plt.scatter(country['Longitude'], country['Latitude']) # plotting the data
plt.xlim(-180,180)
plt.ylim(-90,90)
```

```python
x = country.iloc[:,1:3]  # Selecting the features
```

```python
x
```

```python
# Clustering
```

```python
kmeans = KMeans(2)  # kmeans = KMeans(<number of clusters>)
```

```python
kmeans.fit(x) # kmeans.fit(<dataframe>)
```

```python
identified_clusters = kmeans.fit_predict (x) # extract the results. 
identified_clusters
```

```python
data_with_clusters = country.copy() #copy of the object
```

```python
data_with_clusters
```

```python
data_with_clusters['cluster'] = identified_clusters
```

```python
data_with_clusters
```

```python
plt.scatter(data_with_clusters['Longitude'],data_with_clusters['Latitude'],c = data_with_clusters ['cluster'], cmap ='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show
```

```python
# Obtaining the optimal number of clusters
wcss=[]
for i in range (1,5):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
    
    number_clusters = range(1,5)
```

```python
wcss
```

```python
# Plot the WCSS values

plt.plot (number_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster sum of squares')
```
